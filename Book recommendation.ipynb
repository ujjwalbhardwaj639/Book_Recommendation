## Popularity based recommendation system

```python
ratings.merge(books,on='ISBN').shape
```

```python
book_rating = ratings.merge(books,on='ISBN')
book_rating
```

```python
num_rating = book_rating.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating':'num_rating'},inplace=True)
num_rating
```

```python
avg_rating = book_rating.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating = avg_rating.round(2)
avg_rating
```

```python
trending_df = num_rating.merge(avg_rating,on='Book-Title')
trending_df
```

```python
trending_df[trending_df['num_rating']>=250]
```

```python
trending_df[trending_df['num_rating']>=250].sort_values('avg_rating',ascending=False)
```

```python
trending_df = trending_df[trending_df['num_rating']>=250].sort_values('avg_rating',ascending=False).head(10)
trending_df
```

```python
trending_df.merge(books,on='Book-Title')
multiple books hai wth different isbn number so use command 
```

```python
trending_df.merge(books,on='Book-Title').drop_duplicates('Book-Title').shape
```

```python
trending_df['Image-URL-M'][0]
```

```python
trending_df = trending_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_rating','avg_rating']]
trending_df
```

```python
#Export trending data
import pickle
pickle.dump(trending_df,open('trending.pkl','wb'))
```

## Collaborative filtering based recommendation system

```python
book_rating
```

```python

book_rating.groupby('User-ID').count()['Book-Rating']
```

```python
x = book_rating.groupby('User-ID').count()['Book-Rating']>200
educ_users = x[x].index   
```

```python
book_rating['User-ID'].isin(educ_users) 
```

```python
filtered_rating = book_rating[book_rating['User-ID'].isin(educ_users)]
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scrore = cosine_similarity(pt)
similarity_scrore 
```

```python
similarity_scrore[0]
```

```python
np.where(pt.index=='1984')[0][0]
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(pt.index==book_name)[0][0]
    #distances = similarity_scrore[index]
    similar_items = sorted(list(enumerate(similarity_scrore[index])),key= lambda x:x[1],reverse =True)[1:6]
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(pt.index==book_name)[0][0]
    #distances = similarity_scrore[index]
    similar_items = sorted(list(enumerate(similarity_scrore[index])),key= lambda x:x[1],reverse =True)[1:6]
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    return data
```
